### Imports

```python
from kub.course.plotlib.weatherPlotFactory import WeatherPlotFactory
import pandas as pd

database = "../../database/day2/"
```

### **Weather data analysis**

```python
# Load Weather Data
df_weather = pd.read_csv(database + "Strasbourg/weather-data-0.hourly-variables.csv")
df_weather['time'] = pd.to_datetime(df_weather['time'], unit='ms')
df_weather.set_index('time', inplace=True)

# Use the Weather Factory
weather_factory = WeatherPlotFactory()

# Plot Daily Aggregations
weather_factory.plot_yearly_temperature(df_weather, mode='hourly')
weather_factory.plot_yearly_solar_radiation(df_weather, mode='monthly')
weather_factory.plot_yearly_humidity(df_weather, mode='daily')
```
