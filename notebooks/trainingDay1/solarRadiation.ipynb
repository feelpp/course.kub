### **Exercise 6: Isotropic vs Anisotropic**

**<ins>Question 1:<ins>**

Observe and comment on short wave radiation for south wall displayed for both models.

**<ins>Question 2:<ins>**

Modify **Azimuth** for both model between **0°** and **270°** and **observe change in diffuse radiation for Perez model**.

```python
# Isotropic model
simRadIsotropic = FMUSimulation(database + "Exercices_SolarRadiation_5.fmu")
simRadIsotropic.initialize(startTime=0.0, stopTime=86400.0, timeStep=3600.0)
simRadIsotropic.initParameters( { "building.wall_SOUTH.angleDegAzi": 0.0,
                                  "building.wall_SOUTH.angleDegTil": 90.0 } )
simRadIsotropic.exitInitialization()
```

```python
# Perez model
simRadPerez = FMUSimulation(database + "Exercices_SolarRadiation_6.fmu")

simRadPerez.initialize(startTime=0.0, stopTime=86400.0, timeStep=3600.0)
simRadPerez.initParameters( { "building.wall_SOUTH.angleDegAzi": 0.0,
                              "building.wall_SOUTH.angleDegTil": 90.0 } )
simRadPerez.exitInitialization()
```

```python
# Simulation and plots
outputNames = [ "building.surfacesToAmbience.toSurfacesPorts[2].heatPortSw.Q_flow",
                "building.surfacesToAmbience.toSurfacesPorts[2].radiationPort_in.IrrDir",
                "building.surfacesToAmbience.toSurfacesPorts[2].radiationPort_in.IrrDif" ]

dataIso = simRadIsotropic.run( outputNames )
dataPer = simRadPerez.run( outputNames )

simFactory = SimulationPlotFactory()

simFactory.plot_multi_curves(
    time=dataIso["time"],
    data_type="heat_flux",
    data_dict={ "Short wave radiation Isotropic": dataIso["building.surfacesToAmbience.toSurfacesPorts[2].heatPortSw.Q_flow"],
                "Short wave radiation Perez": dataPer["building.surfacesToAmbience.toSurfacesPorts[2].heatPortSw.Q_flow"] },
    title="Exercise 6 — Short wave radiation (Direct + Diffuse)"
)

simFactory.plot_multi_curves(
    time=dataIso["time"],
    data_type="surface_heat_flux",
    data_dict={ "Direct radiation Isotropic": dataIso["building.surfacesToAmbience.toSurfacesPorts[2].radiationPort_in.IrrDir"],
                "Diffuse radiation Isotropic": dataIso["building.surfacesToAmbience.toSurfacesPorts[2].radiationPort_in.IrrDif"],
                "Direct radiation Perez": dataPer["building.surfacesToAmbience.toSurfacesPorts[2].radiationPort_in.IrrDir"],
                "Diffuse radiation Perez": dataPer["building.surfacesToAmbience.toSurfacesPorts[2].radiationPort_in.IrrDif"] },
    title="Exercise 6 — Solar radiation (Direct + Diffuse)"
)
```
